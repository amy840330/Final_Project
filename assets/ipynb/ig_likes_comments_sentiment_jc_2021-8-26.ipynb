```python
# Import Dependencies and setup
import pandas as pd
```

```python
# read csv output from Instagrapy_split_text.ipynb
df=pd.read_csv("../../resources/New_sentiment_IG_Category_jc_2021-08-26.csv", encoding="ISO 8859-1")
df.head(2)
```

```python
# Extract selected columns to a new dataframe with key 
new_df = df.filter(['author','Category','likes', 'comments', 'neg','neu','pos', 'compound'], axis=1)
new_df.head()
```

```python
# Rename columns
new_df.rename(columns = {'author':'Author', 'likes':'Likes', 
                                     'comments':'Comments', 'neg':'Neg', 
                                     'pos':'Pos', 'compound':'Compound'}, inplace = True)

new_df = new_df.reset_index(drop=True)
new_df.head()
```

```python
new_df.to_csv("../../resources/ig_likes_comments_sentiment_jc_2021-08-26")
```
