```python
from instagramy import InstagramUser
import pandas as pd
```

```python
session_id = "145470550%3A7nYpcQwywv4i17%3A15"
```

```python
user_list = [
    'thesmokingtire'
]
```

```python
user_followers = []
user_posts = []

def scrape_detail():
    for x in user_list:
        # Connect to profile
        user = InstagramUser( x, sessionid=session_id)
        # Scarpe specific info
        user_followers.append( user.number_of_followers)
        user_posts.append( user.number_of_posts)
        
scrape_detail()
```

```python
print( user_list)
print( user_followers)
print( user_posts)
```

```python
user_list_df = pd.DataFrame( columns= ['account', 'followers', 'posts'])
```

```python
user_list_df.account = user_list
user_list_df.followers = user_followers
user_list_df.posts = user_posts

user_list_df
```
